## Clean the data

```python
#Changing some data types

df['Id'] = df['Id'].astype(str)
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'], format ='%m/%d/%Y')
df.dtypes
```

```python
#Checking some columns

df['distance_diff'] = df['TotalDistance'] - df['TrackerDistance']
df['distance_diff'].value_counts()
```

```python
#Renaming columns

df.columns = df.columns.str.lower()
df.columns
```

```python
df.rename (columns = { 'activitydate': 'activity_date', 'totalsteps': 'total_steps', 'totaldistance': 'total_distance',
                      'trackerdistance': 'tracker_distance',
       'loggedactivitiesdistance': 'logged_activities_distance', 'veryactivedistance': 'very_actived_distance',
       'moderatelyactivedistance': 'moderately_active_distance', 'lightactivedistance': 'light_active_distance',
       'sedentaryactivedistance': 'sedentary_active_distance', 'veryactiveminutes': 'very_active_minutes',
                      'fairlyactiveminutes': 'fairly_active_minutes',
       'lightlyactiveminutes': 'lightly_active_minutes', 'sedentaryminutes': 'sedentary_minutes', }, inplace=True)
df.columns
```

```python
#Creating new columns

df['day_of_week'] =df['activity_date'].dt.day_name()
df['n_day_of_week']=df['activity_date'].dt.weekday
df.head(10)
```

```python
# Checking for null values and duplicates

print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
df['id'].nunique()
```

## Analysis

#### We will start the analysis phase with grouping the users by their id

```python
id_grp = df.groupby(['id'])
id_avg_step = id_grp['total_steps'].mean().sort_values(ascending=False)
id_avg_step = id_avg_step.to_frame()
```

```python
id_avg_step 
```

### Categorization of every user 

```python
conditions = [
   (id_avg_step <= 6000),
   (id_avg_step > 6000) & (id_avg_step < 12000),
   (id_avg_step >= 12000)
]
values = ['sedentary','active','very_active']

id_avg_step['activity_level'] = np.select(conditions,values)

id_activity_level = id_avg_step['activity_level']
```

```python
id_avg_step
```

```python
df['activity_level'] =  [id_activity_level[c] for c in df['id']]
df.head()
```

```python
df.describe()
```

### We are looking for the correlation between steps and calories burned

```python
ax = sns.scatterplot (x='total_steps', y='calories', data=df, hue=df['activity_level'])

plt.title('Correlation calories vs. steps')

plt.tight_layout()
plt.show()
```

### Now we are going to see the average steps per day

```python
day_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

fig, ax = plt.subplots(1,1, figsize=(8,5))

day_grp = df.groupby(['day_of_week'])
avg_daily_steps = day_grp['total_steps'].mean().reindex(day_of_week)
avg_steps = df['total_steps'].mean()

plt.bar(avg_daily_steps.index, avg_daily_steps, color='blue')
plt.title('Average daily steps per day')
plt.show()
```

### We are also going to show the percentage of activity in minutes

```python
very_active_mins = df['very_active_minutes'].sum()
fairly_active_mins = df['fairly_active_minutes'].sum()
lightly_active_mins = df['lightly_active_minutes'].sum()
sedentary_mins = df['sedentary_minutes'].sum()

slices = [very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins]
labels = ['Very active minutes', 'Fairly active minutes', 'Lightly active minutes', 'Sedentary minutes']
explode = [0,0,0,0.1]

plt.pie(slices, labels=labels, explode=explode, autopct= '%1.1f%%')
plt.title('Percentage of activity level in minutes')
plt.show()
```
